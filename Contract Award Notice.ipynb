```python
import pandas as pd

table_CAN = pd.read_html('https://www.ha.org.hk/haho/ho/bssd/TA_236491_210026111a.htm')
```

```python
print(f'Total tables: {len(table_CAN)}')
```

```python
df = table_CAN[0]
df.head(20)
```

```python
df.drop(df.index[0:6], inplace=True)
```

```python
df.drop(columns=['index'], inplace=True, errors='ignore')
```

```python
df = df.drop_duplicates()
df
```

```python
# Set the first as header
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header
```

```python
df.info()
```

```python
df['Hospital'] = df['Hospital'].astype(str)
```

```python
df = df[~df['Hospital'].str.contains('Awarded Month')]
```

```python
df = df[~df['Hospital'].str.contains('nan')]
```

```python
filtered_df = df[df['Contractor(s) & Address(es)'].str.contains('dksh', case=False)]
```

```python
filtered_df.head()
```

```python
filtered_df.to_excel('filter_CAN.xlsx',sheet_name='Data')
```

```python
df.to_excel('CAN.xlsx',sheet_name='Data')
```
